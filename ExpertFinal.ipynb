```python
from experta import *

def delete():
  f = open("Security.txt", "w")
  f.close() 

def write(a):
  f = open("Security.txt", "a")
  f.write(a)
  f.close() 

def read():
  f = open("Security.txt", "r")
  print(f.read())
```

```python
class Security(KnowledgeEngine):
  @DefFacts()
  def _initial_action(self):
    yield Fact(action="start")

  @Rule(Fact(action="start"),
        NOT(Fact(budget=W())))
  def select_budget(self):
    self.declare(Fact(budget=input("Do you want high budget security?(y,n) ")))

#Low Budget Security

  @Rule(Fact(action='start'),
        Fact(budget='n'))
  def kkkkk(self):
      write("\nLow Budget Security : \n")
      self.declare(Fact(high_level_security=input("Do you want high level security?(y,n)")))

  @Rule(Fact(action='start'),
        Fact(high_level_security='n'))
  def ask_level_out(self):
      write("-Company door: securityguard, Security cam, Intrusion Alarm \n-Secure wireless network \n")
    
  @Rule(Fact(action='start'),
        Fact(high_level_security='y'))
  def ask_data(self):
      self.declare(Fact(data=input("Do you have valuable data?(y,n)")))
      write("\n-Company door: securityguard, Security cam , Intrusion Alarm \n-Secure wireless network \n")

#valuable_data  
  @Rule(Fact(action='start'),
        Fact(data='y'))
  def ask_nettwork(self):
      self.declare(Fact(nettwork=input("Do you have network?(y,n)")))
      write("-Backup to cloud \n-Install antivirus software \n-Set up automatic software updates \n")

  @Rule(Fact(action='start'),
        Fact(data='n'))
  def ask_nettwork1(self):
      self.declare(Fact(nettwork=input("Do you have network?(y,n)")))
#network
  @Rule(Fact(action='start'),
        Fact(nettwork='y'))
  def ask_Sroom(self):
      self.declare(Fact(Sroom=input("Do you have server room ?(y,n)")))
      write("-Apply firewall \n-Apply VPN \n")
    
    
#Sroom    
  @Rule(Fact(action='start'),
        Fact(Sroom='y'))
  def ask_Sroom_out(self):
      write("-Server room: no windows , door opens only to the server admin and the owner of the company \n")
    
   
#High Budget Security


  @Rule(Fact(action="start"),
        Fact(budget='y'),
        NOT(Fact(choice=W())))
  def select_type(self):
    self.declare(Fact(choice=input("What type of security do you want(Data,Network,Building)?(d,n,b) ")))

  @Rule(Fact(action="start"),
        Fact(choice=MATCH.choice))
  def svs(self,choice):
    if 'n' in choice:
      self.declare(Fact(nchoice='y'))

    if 'b' in choice:
      self.declare(Fact(bchoice='y'))

    if 'd' in choice:
      self.declare(Fact(dchoice='y'))


#Network

  @Rule(Fact(action='start'),
        NOT(Fact(server=W())),
        Fact(nchoice='y'))
  def ask_server(self):
    print("\n")
    print("Network Questions: \n")
    write("\n\nNetwork Security : \n")
    self.declare(Fact(server=input("Do you have server?(y,n)")))

  #server
  @Rule(Fact(action='start'),
         Fact(server='y'))
  def ask_server_room(self):
      self.declare(Fact(server_room=input("Is it in a server room?(y,n)"))) 
      
  #server room
  @Rule(Fact(action='start'),
        Fact(server_room='y'))
  def ask_server_room1(self):
      #output
      write("Your server room must have no windows to open.\nServer room must be safe from water lacking.\nYou must have high security door lock.\n")
   
  #server room
  @Rule(Fact(action='start'),
        Fact(server_room='n'))
  def ask_locker(self):
      self.declare(Fact(server_room_locker=input("Can you store the devices in one locker?(y,n)"))) 
    
  #locker
  @Rule(Fact(action='start'),
        Fact(server_room_locker='y'))
  def ask_locker1(self):
      write("Put your devices in a locker and keep it cool.\n")
    
  #locker
  @Rule(Fact(action='start'),
        Fact(server_room_locker='n'))
  def ask_locker2(self):
      write("You need a server room .\n")
        
  #server
  @Rule(Fact(action='start'),
        OR(Fact(server='n'),Fact(server_room_locker=W()),Fact(server_room='y')))
  def ask_firewall(self):
      self.declare(Fact(firewall=input("Do you need a firewall?(y,n)"))) 
         
  #firewall
  @Rule(Fact(action='start'),
        Fact(firewall='y'))
  def ask_firewall_quality(self):
      self.declare(Fact(firewall_quality=input("High quality firewall?(y,n)"))) 
        
  #firewall
  @Rule(Fact(action='start'),
        OR(Fact(firewall='n'),Fact(firewall_quality=W())))
  def ask_VPN(self):
      self.declare(Fact(VPN=input("You need a VPN?(y,n)"))) 
    
  #VPN
  @Rule(Fact(action='start'),
        Fact(VPN='y'))
  def ask_VPN_quality(self):
      self.declare(Fact(VPN_quality=input("High quality  VPN?(y,n)"))) 
    
  #VPN
  @Rule(Fact(action='start'),
        AND(Fact(VPN='n'),Fact(firewall='n')))
  def ask_VPN_out(self):
      write("Unsafe data.\n You need VPN and/or firewall \n")
    
  #VPN quality
  @Rule(Fact(action='start'),
        Fact(VPN_quality='y'))
  def ask_VPN_quality_out(self):
      write("High quality VPN .\n")

  @Rule(Fact(action='start'),
        Fact(VPN_quality='n'))
  def ask_VPN_quality_out1(self):
      write("Normal quality VPN .\n")
        
  #firewall quality
  @Rule(Fact(action='start'),
        Fact(firewall_quality='y'))
  def ask_firewall_quality_out(self):
      write("High quality firewall .\n")
  
  @Rule(Fact(action='start'),
        Fact(firewall_quality='n'))
  def ask_firewall_quality_out1(self):
      write("Normal quality firewall .\n")



#########################################################################################

#Building:

  @Rule(Fact(action="start"),
        Fact(bchoice='y'))
  def b1(self):
    print("\n")
    print("Building Questions: \n")
    self.declare(Fact(v1=input("Do you want a security team? (y,n)")))
    write("\n\nBuilding Security : \n")

  @Rule(Fact(action="start"),
        Fact(v1='y'))
  def b2(self):
    self.declare(Fact(v2=input("Do you have one building or multiple buildings?(o,m)")))

  @Rule(Fact(action="start"),
        Fact(v2='o'))
  def b3(self):
    write("Add small number of security guards(2..5).\n")

  @Rule(Fact(action="start"),
        Fact(v2='m'))
  def b4(self):
    write("Add huge number of security guards(10+).\n")

  @Rule(Fact(action="start"),
        (OR(Fact(v1='n'),Fact(v2=W()))))
  def b5(self):
    self.declare(Fact(v3=input("Does your employees need protection?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v3='y'))
  def b6(self):
    write("Add GPS tracking system for employees.\n")

  @Rule(Fact(action="start"),
        Fact(v3=W()))
  def b7(self):
    self.declare(Fact(v4=input("Do you have a storehouse?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v4='y'))
  def b8(self):
    write("Install CCTV cameras at the storehouse.\n")

  @Rule(Fact(action="start"),
        Fact(v4=W()))
  def b9(self):
    self.declare(Fact(v5=input("Do you need a security checkpoint?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v5='y'))
  def b10(self):
    write("Add security checkpoint.\n")
    self.declare(Fact(v6=input("Do you need a weapons checkpoint?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v6='y'))
  def b11(self):
    write("Add weapons checkpoint.\n")


  @Rule(Fact(action="start"),
        Fact(v6=W()))
  def b12(self):
    self.declare(Fact(v7=input("Do you need a bags checkpoint?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v7='y'))
  def b13(self):
    write("Add bags checkpoint.\n")

  @Rule(Fact(action="start"),
        OR(Fact(v5='n'),Fact(v7=W())))
  def b14(self):
    self.declare(Fact(v8=input("Do you need fire peotection system?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v8='y'))
  def b15(self):
    write("Add fire protection system.\n")

  @Rule(Fact(action="start"),
        Fact(v8=W()))
  def b16(self):
    self.declare(Fact(v9=input("Do you need intrusion detection system?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v9='y'))
  def b17(self):
    write("Add intrusion detection system.\n")

  @Rule(Fact(action="start"),
        Fact(v9=W()))
  def b18(self):
    self.declare(Fact(v10=input("Do you want cameras installed outside the building?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v10='y'))
  def b19(self):
    write("Install cameras outside the building.\n")

  @Rule(Fact(action="start"),
        Fact(v10=W()))
  def b20(self):
    self.declare(Fact(v11=input("Do you want physical barriers installed outside the building?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v11='y'))
  def b21(self):
    write("Install physical barriers outside the building.\n")

  @Rule(Fact(action="start"),
        Fact(v11=W()))
  def b22(self):
    self.declare(Fact(v12=input("Do you want garage protection system?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v12='y'))
  def b23(self):
    write("Install CCTV cameras, alarms, sensors in the garage.\n")

  @Rule(Fact(action="start"),
        Fact(v11=W()))
  def b22(self):
    self.declare(Fact(v12=input("Do you want garage protection system?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v12='y'))
  def b23(self):
    write("Install CCTV cameras, alarms, sensors in the garage.\n")

  @Rule(Fact(action="start"),
        Fact(v12=W()))
  def b24(self):
    self.declare(Fact(v13=input("Do you have a money deposit box room?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v13='y'))
  def b25(self):
    write("Install heavy doors and alarms at money deposit box room.\n")

  @Rule(Fact(action="start"),
        Fact(v13=W()))
  def b26(self):
    self.declare(Fact(v14=input("Do you want secure entry system for customers?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v14='y'))
  def b27(self):
    write("Initialize customer entry registration system.\n")

  @Rule(Fact(action="start"),
        Fact(v14=W()))
  def b28(self):
    self.declare(Fact(v15=input("Do you have regular cargo transportaion in and out of the building?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v15='y'))
  def b29(self):
    self.declare(Fact(v16=input("Do you want full cargo transportation security system?(y,n)")))

  @Rule(Fact(action="start"),
        Fact(v16='y'))
  def b30(self):
    write("Add real time tracking system for cargo trucks.\n")

#########################################################################################

#Data:

  @Rule(Fact(action='start'),
        Fact(dchoice='y'))
  def d0(self):
    print("\n")
    print("Data Questions: \n")
    write("\n\nData Security: ")
    self.declare(Fact(xbudget=input("Do you want high budget antivirus?(y/n) ")))

  @Rule(Fact(action='start'),
        Fact(xbudget='y'))
  def d1(self):
    write("\nAdd high budget antivirus ")

  @Rule(Fact(action='start'),
        Fact(xbudget=W()))
  def d2(self):
    self.declare(Fact(xemp=input("Do you need employee training?(y/n) ")))

  @Rule(Fact(action='start'),
        Fact(xemp='y'))
  def d3(self):
    write("\nEnroll your employees in data security training program ")

  @Rule(Fact(action='start'),
        Fact(xemp=W()))
  def d4(self):
    self.declare(Fact(xpol=input("Do you need a policy to your company?(y/n) ")))

  @Rule(Fact(action='start'),
        Fact(xpol='y'))
  def d5(self):
    write("\nInitialize a data protection policy in your company ")

  @Rule(Fact(action='start'),
        Fact(xpol=W()))
  def d6(self):
    self.declare(Fact(xcyber=input("Do you need CyberSecurity ?(y/n) ")))

  @Rule(Fact(action='start'),
        Fact(xcyber='y'))
  def d7(self):
    write("\nHire Employees specialized in CyberSecurity")



```
